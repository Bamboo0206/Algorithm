```python
from geopy.distance import geodesic
import numpy as np
```

```python
# data=np.ndarray([])
data=np.loadtxt('data_28.txt')
data
```

```python
result=np.zeros((data.shape[0],data.shape[0]))
for i in range(data.shape[0]):
    for j in range(data.shape[0]):
        if(i==j):
            result[i][j]=0
        else:
            result[i][j]=geodesic(data[i], data[j]).m
            
            
result
```

```python
np.savetxt('dist_matrix.txt',result,delimiter='\t',fmt="%.16f")
```
